**Objective**: Use this notebook to quickly test the classification of new financial transaction descriptions using your trained models.

## 1. Setup and Load Model

```python
import os
import sys

# Add src to path to use project utilities
sys.path.append(os.path.abspath(".."))
from src.models.classifier import FinancialClassifier
from src.data.preprocessing import preprocess_text

# 📂 Path to your best model
MODEL_PATH = "../models/LinearSVM_50_50.joblib"

if os.path.exists(MODEL_PATH):
    # Correct way to load a classmethod that returns an instance
    model = FinancialClassifier.load(MODEL_PATH)
    print(f"✅ Model loaded successfully from: {MODEL_PATH}")
else:
    print(f"❌ ERROR: Model file not found at {MODEL_PATH}. Please train a model first.")
```

## 2. Define Prediction Logic

```python
def classify_description(text):
    """
    Preprocesses the input text and returns the predicted area.
    """
    # 1. Clean the text using our project's logic
    cleaned_text = preprocess_text(text)
    
    # 2. Predict
    prediction = model.predict([cleaned_text])[0]
    
    return cleaned_text, prediction

print("💡 Function ready for use!")
```

## 3. Test your own descriptions

```python
my_descriptions = [
    "Cena restaurante italiano con amigos",
    "Ingreso de nómina mensual empresa SL",
    "Compra de acciones Apple Inc.",
    "Recibo de luz mes de enero",
    "Entradas para el cine Warner",
    "Transferencia recibida de mi tía"
]

print(f" {'DESCRIPTION':<40} | {'PREDICTION':<15} ")
print("-" * 60)

for desc in my_descriptions:
    _, category = classify_description(desc)
    print(f" {desc:<40} | {category:<15} ")
```

## 4. Interactive Mode

```python
user_input = input("✍️ Enter a transaction description: ")
while(user_input != ""):
    clean, cat = classify_description(user_input)
    print(f"\n--- Result ---")
    print(f"Original: {user_input}   |   Processed: {clean}")
    print(f"Predicted Area: {cat}")
    user_input = input("✍️ Enter a transaction description: ")
    
```
