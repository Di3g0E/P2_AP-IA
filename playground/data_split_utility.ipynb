# Utilidad de División de Datos (Split Dataset)

Este notebook permite dividir un conjunto de datos en entrenamiento (train) y prueba (test) con una proporción **50/50**, manteniendo la estratificación por la columna `Area`. Los archivos resultantes se guardan en la carpeta `data/processed/` con nombres que indican su origen.

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from pathlib import Path

# Configuración de rutas
BASE_DIR = Path("..")
RAW_DATA_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DATA_DIR = BASE_DIR / "data" / "processed"

# Asegurar que la carpeta de salida existe
os.makedirs(PROCESSED_DATA_DIR, exist_ok=True)

def split_and_save_dataset(filename, filename2=None, target_column='Area', test_size=0.5):
    """
    Carga un dataset, lo divide en train/test y guarda los resultados.
    """
    file_path = RAW_DATA_DIR / filename
    if not file_path.exists():
        print(f"Error: El archivo {filename} no existe en {RAW_DATA_DIR}")
        return

    if filename2:
        file_path2 = RAW_DATA_DIR / filename2
        if not file_path2.exists():
            print(f"Error: El archivo {filename2} no existe en {RAW_DATA_DIR}")
            return
    
    # Cargar datos
    df = pd.read_csv(file_path)
    print(f"Dataset cargado: {filename} ({len(df)} filas)")

    if filename2:
        df2 = pd.read_csv(file_path2)
        print(f"Dataset cargado: {filename2} ({len(df2)} filas)")

        df_resultado = pd.concat([df, df2], ignore_index=True)
    
    # Limpiar posibles nulos en la columna objetivo para el split
    df = df.dropna(subset=[target_column])
    
    # División estratificada 50-50
    train_df, test_df = train_test_split(
        df, 
        test_size=test_size, 
        random_state=42, 
        stratify=df[target_column]
    )
    
    # Construir nombres de archivo
    base_name = Path(filename).stem
    train_output = PROCESSED_DATA_DIR / f"{base_name}_train.csv"
    test_output = PROCESSED_DATA_DIR / f"{base_name}_test.csv"
    
    # Guardar CSVs
    train_df.to_csv(train_output, index=False)
    test_df.to_csv(test_output, index=False)
    
    print(f"✅ Archivos guardados en {PROCESSED_DATA_DIR}:")
    print(f"   - {train_output.name} ({len(train_df)} filas)")
    print(f"   - {test_output.name} ({len(test_df)} filas)")
```

## Ejecución de la división
Puedes cambiar el nombre del archivo abajo para procesar cualquiera de los datasets en `data/raw/`.

```python
# Lista de archivos disponibles en data/raw para referencia
print("Archivos disponibles:", os.listdir(RAW_DATA_DIR))

# Ejemplo con db_orig.csv
split_and_save_dataset("db_mod_descript.csv")
```
